```python
#there are null value in the column as it is numeric data we will replace it by mean 
df["retail_price"] = df["retail_price"].fillna(df["retail_price"].mean())
```

```python
#there are null value in the column as it is numeric data we will replace it by mean
df["discounted_price"] = df["discounted_price"].fillna(df["discounted_price"].mean())
```

```python
#there are null value in the column as it is categorical data we will replace it by mode
df["brand"] = df["brand"].fillna(df["brand"].mode().iloc[0])
```

```python
df.info()
```

```python
df.isnull().sum() 
```

```python
# still there are null value which is less that 3% so we will drop that rows using dropna
df.dropna(inplace=True)
```
